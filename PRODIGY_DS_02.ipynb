#Data Cleaning

```python
dataset['Age'].fillna(dataset['Age'].median(), inplace=True)
```

#Exploratory Data Analysis

```python
import seaborn as sns
```

# Histogram of Age


```python
sns.histplot(dataset['Age'], bins=20, kde=True, color="blue")
```

```python
dataset.hist(bins=20, figsize=(14, 10), layout=(3, 3),color="red")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(dataset['Age'], bins=20, kde=True, color="orange")
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
print(dataset.describe(include='all'))
```

```python
# Boxplot of Fare
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Fare', data=dataset)
plt.title("Survival by Fare")
plt.xlabel("Survived")
plt.ylabel("Fare")
plt.show()
```

```python
#Explore realtionship between variables
```

```python
#Survival Rate by Passenger Class
sns.countplot(x='Pclass', hue='Survived', data=dataset)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
#Survival Rate by Gender
sns.countplot(x='Sex', hue='Survived', data=dataset)
plt.title('Survival by Gender')
plt.show()
```

```python
#Survival Rate by Age
# Creating a facet grid for better visualization
g = sns.FacetGrid(dataset, col='Survived')
g.map(plt.hist, 'Age', bins=20)
plt.show()
```

```python
#Survival Rate by Fare
# Creating a facet grid for better visualization
g = sns.FacetGrid(dataset, col='Survived')
g.map(plt.hist, 'Fare', bins=20)
plt.show()
```

```python
#Correlations Between Numerical Features
# Convert 'Survived' to int for correlation calculation
dataset['Survived'] = dataset['Survived'].astype(int)

# Compute the correlation matrix
correlation_matrix = dataset[['Survived', 'Age', 'Fare']].corr()

# Display the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#Pairplot of Numerical Features
sns.pairplot(dataset[['Survived', 'Age', 'Fare']], hue='Survived')
plt.show()
```

```python
#Boxplot of Fare by Class
sns.boxplot(x='Pclass', y='Fare', data=dataset)
plt.title('Fare Distribution by Class')
plt.show()
```
